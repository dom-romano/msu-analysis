#### Overall

Let's start by analyzing efficiency in scoring by calculating points per shot attempt.

```python
df['Points Per Shot'] = df['PTS'] / df['FGA']
sorted_pps = df.sort_values(by='Points Per Shot', ascending=False)

sorted_pps
```

Now let's move onto rebounding. We'll analyze player rebounds as a rate compared to total team rebounds.

```python
# Calulate team total rebounds
team_total_rebounds = df['TRB'].sum()
team_total_rebounds

# Calulate player rebound rate
df['Rebound Rate'] = df['TRB'] / team_total_rebounds

# Sort list
sorted_rr = df.sort_values(by='Rebound Rate', ascending=False)

sorted_rr
```

```python
# Pie chart based on sorted_rr
import matplotlib.pyplot as plt
plt.style.use('ggplot')
explode_king = [0.1 if player == 'Blessing King' else 0 for player in sorted_rr['Player']]
plt.pie(sorted_rr['Rebound Rate'], labels=sorted_rr['Player'], autopct='%1.1f%%', explode=explode_king)
plt.show()
```

Again, Blessing King tops this list with a staggering 26% of the teams total rebounds. Hallie Rhodes and Katie Novik take the number 2 and 3 spots, respectively.

Now, let's turn to assist-to-turnover ratios. This will tell us who creates consistent scoring opportunities with minimal turnovers.

```python
df['Assist to Turnover Ratio'] = df['AST'] / df['TOV']

sorted_atr = df.sort_values(by='Assist to Turnover Ratio', ascending=False)

sorted_atr
```

```python
sorted_atr.plot(x='Player', y='Assist to Turnover Ratio', kind='bar')
```

Veronica Charles takes the top spot on this list with an average of 1.28 assists per turnover. Katie Novik takes the 2nd spot with 1.09.


Now let's take a look at each player's contributions to games. We will assign a weight to each stat and give a score to each player that determines their contributions to each game.

```python
# Assign weights

weights = {
    "PTS" : 0.5,
    "AST" : 0.3,
    "TRB" : 0.2,
    "STL" : 0.15,
    "BLK" : 0.1,
    "TOV" : -0.2,
    "3P" : 0.6
}

# Normalize each player stat by team average
df['PTS_norm'] = df['PTS'] / df['PTS'].mean()
df['AST_norm'] = df['AST'] / df['AST'].mean()
df['TRB_norm'] = df['TRB'] / df['TRB'].mean()
df['STL_norm'] = df['STL'] / df['STL'].mean()
df['BLK_norm'] = df['BLK'] / df['BLK'].mean()
df['TOV_norm'] = df['TOV'] / df['TOV'].mean()
df['3P_norm'] = df['3P'] / df['3P'].mean()

# Calculate player scores
df['Contribution Score'] = df.apply(lambda row: sum(row[stat] * weights[stat] for stat in weights), axis=1)

# Sort list
sorted_cs = df.sort_values(by='Contribution Score', ascending=False)

sorted_cs
```

```python
# Plot contribution score
sorted_cs.plot(x='Player', y='Contribution Score', kind='bar')
```

Veronica Charles has the highest contribution score, meaning she has the most positive impact on **both** sides of the court.

#### Offensive

Let's get a little more specific and talk about offense. We will create offensive contribution scores, much like the overall contribution score above. We'll be using the per-minute stats to get a more specific score.

```python
# Calculate offensive stats per minute
df['Points Per Minute'] = df['PTS'] / (df['MP'])
df['Assists Per Minute'] = df['AST'] / (df['MP'])
df['3-Point Field Goals Per Minute'] = df['3P'] / (df['MP'])

# Assign weights
weights = {
    'Points Per Minute' : 0.4,
    'Assists Per Minute' : 0.3,
    '3-Point Field Goals Per Minute' : 0.2,
    'FG%' : 0.1,
    'FT%' : 0.1
}

# Normalize each player stat by team average of each
df['PPM_norm'] = df['Points Per Minute']/ df['Points Per Minute'].mean()
df['APM_norm'] = df['Assists Per Minute']/ df['Assists Per Minute'].mean()
df['3PMPM_norm'] = df['3-Point Field Goals Per Minute']/ df['3-Point Field Goals Per Minute'].mean()
df['FG%_norm'] = df['FG%'] / df['FG%'].mean()
df['FT%_norm'] = df['FT%'] / df['FT%'].mean()

# Calculate offensive contribution score
df['Offensive Contribution Score (Per Minute)'] = df.apply(lambda row: sum(row[stat] * weights[stat] for stat in weights), axis=1)

# Sort list
sorted_ocs = df.sort_values(by='Offensive Contribution Score (Per Minute)', ascending=False)
sorted_ocs
```

```python
sorted_ocs.plot(x='Player', y='Offensive Contribution Score (Per Minute)', kind='bar')
```

This graph shows that at any given minute, on offense, Valentina Saric is contributing the most, on average.

Now, let's analyze each players scoring impact per possession.

To do this, we will estimate points per possession using a formula that combines their points per game with an estimate of how often they are involved in the offense.

```python
# Define variables that will be used in formula
ppg = df['PTS']
fga = df['FGA']
fta = df['FTA']
to = df['TOV']

# Calculate points per possession
# 0.44 is a widely used constant in basketball analytics to account for free throw attempts
ppp = ppg/(0.44*fta+fga+to)

# Add a points per possession column
df['Points Per Possession'] = ppp

# Sort list
sorted_ppp = df[['Player', 'Points Per Possession']].sort_values(by='Points Per Possession', ascending=False)
sorted_ppp
```

According to this list, Valentina Saric scores 1.11 points per possession. This metric indicates how efficiently a player scores on each possession.

#### Defensive

Let's calculate a few scores based on defensive stats.

* Defensive Play Impact: A weighted score giving steals the most weight.

* Possession Disruption Rate: How often a player disrupts the opponent's possession.

* Defensive Clutch Impact: Calculates a player's defensive efficiency per minute

```python
# Defensive Play Impact
df['DPI'] = (df['STL'] * 3) + (df['BLK'] * 2) + (df['DRB'])

df[['Player', 'DPI']].sort_values(by='DPI', ascending=False)
```

```python
# Possession Disruption Rate
# This formula uses the widely used constant of 0.44 to account for free throw attempts

df['PDR'] = (df['STL'] + df['BLK']) / (0.44 * df['FTA'] + df['FGA'] + df['TOV'])

df[['Player', 'PDR']].sort_values(by='PDR', ascending=False)
```

```python
# Defensive Clutch Impact
df['DCI'] = (df['STL'] + df['BLK']) / (df['MP'])

df[['Player', 'DCI']].sort_values(by='DCI', ascending=False)
```

Now let's make a normalized defensive score, taking these stats into consideration.

```python
defensive_score = (df['DPI'] + df['PDR'] + df['DCI']) / 3

df['Defensive Score'] = defensive_score

df[['Player', 'Defensive Score']].sort_values(by='Defensive Score', ascending=False)
```

The top defensive player is Blessing King. Her strength is disrupting opponent's possession. She also is the leader in defensive play impact, indicating an elite defensive player.

In high-stakes defensive situations, consider Blessing King.

Now, let's calculate who is the most effective at contesting shots.

While blocks are the most obvious form of shot contesting, many great defenders alter shots without actually blocking them. Let's use a shot contestation metric dervied from blocks, steals and defensive rebounds.

```python
# Calculate shot contestation impact
df['Shot Contestation Impact'] = (df['BLK'] * 1.5) + (df['STL']) + (df['DRB'] * 0.5)

df[['Player', 'Shot Contestation Impact']].sort_values(by='Shot Contestation Impact', ascending=False)
```

Again, Blessing King tops this list. She is the best at contesting shots.
